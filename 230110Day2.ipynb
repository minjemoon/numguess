```python
animals = ('rabbit','cat','dog','aligator','tiger','lion')
animals = list(animals)
animals.append('hamster')
animals.append('elephant')
animals.append('rhino')
print(animals)
print(type(animals))
animals = tuple(animals)
print(animals)
print(type(animals))
sorted_animals = sorted(animals, reverse=True)
print(sorted_animals)
even_animals = animals[2::2]
print(even_animals)
```
